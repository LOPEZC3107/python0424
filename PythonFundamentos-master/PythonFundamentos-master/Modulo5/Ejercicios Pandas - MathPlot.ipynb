```python
import os
import pandas as pd
```

```python
df = pd.read_csv('./src/input/titanic.csv')
df.head(2)
```

```python
# Realizar ejercicios 3 y 8
```

```python
condicion = df['PassengerId']==148
df[condicion]
```

```python
# Eliminar del DataFrame los pasajeros con edad desconocida.
# https://pandas.pydata.org/docs/getting_started/intro_tutorials/03_subset_data.html#min-tut-03-subset

condicion = df['Age'].notna() # Nan representan valores vacios en mi df
df_nuevo = df[condicion]
df_nuevo
```

```python
# 5. Mostrar por pantalla los nombres de las personas que iban en primera clase ordenadas alfabéticamente.
df.head(2)
```

```python
df['Pclass'].unique()
```

```python
df_first_class = df[df['Pclass']==1]
df_first_class.sort_values(by='Name',ascending=True)
```

```python
# 8. Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.

df_mujeres = df[df['Sex']=='female']

df_mujeres.groupby('Pclass').agg({
    'Age':['min','mean', 'max', 'count', 'sum']
    
})
```

```python
# 9. Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no.
```

```python
def etiquetar_por_edad(edad):
    if edad>=18:
        return 'Mayor edad'
    elif edad<18:
        return "Menor edad"
```

```python
df['etiqueta_edad'] = df['Age'].apply(etiquetar_por_edad)
```

```python
df
```

```python
# Supongamos que te piden realizar un reporte bajo el cual requieres mostrar de manera separada información de cada sexo y clase en un archivo excel
# Debemos generar por sexo y clase un reporte que contenga los datos de los pasajeros
```

```python
df_sex_class = df[['Sex', 'Pclass']].drop_duplicates(subset=['Sex', 'Pclass'])

df_sex_class
```

```python
for sex in df.Sex.unique():
    df_sex = df[df['Sex']==sex]
    
    for clase in df.Pclass.unique():
        df_sex_class = df_sex[df_sex.Pclass==clase] 
        
        filename = f'Reporte-{sex}-clase{clase}'
        df_sex_class.to_excel(f'src/output/reporte_titanic/{filename}.xlsx', index=False, sheet_name=filename)
```
